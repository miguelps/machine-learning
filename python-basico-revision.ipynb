## 4. Listas y Diccionarios


```python
# Listas
numeros = [1, 2, 3, 4, 5]
frutas = ["manzana", "banana", "naranja"]

print(f"Lista de n칰meros: {numeros}")
print(f"Primer elemento: {numeros[0]}")
print(f"칔ltimo elemento: {numeros[-1]}")
print(f"Slicing [1:3]: {numeros[1:3]}")

# Agregar elementos
frutas.append("uva")
print(f"\nDespu칠s de append('uva'): {frutas}")

# M칠todos 칰tiles
print(f"\nLongitud: {len(frutas)}")
print(f"Ordenado: {sorted(frutas)}")
```

```python
# Diccionarios
persona = {
    "nombre": "Miguel",
    "edad": 30,
    "ciudad": "Arequipa"
}

print(f"Diccionario completo: {persona}")
print(f"Nombre: {persona['nombre']}")
print(f"Edad: {persona.get('edad')}")
print(f"Email (con default): {persona.get('email', 'N/A')}")

# Iterar
print("\nIterando sobre el diccionario:")
for clave, valor in persona.items():
    print(f"  {clave}: {valor}")
```

## 5. Funciones


```python
# Funci칩n simple
def saludar(nombre):
    return f"춰Hola, {nombre}!"

print(saludar("Miguel"))

# Funci칩n con valor por defecto
def saludar_completo(nombre, mensaje="Bienvenido"):
    return f"{mensaje}, {nombre}!"

print(saludar_completo("Miguel"))
print(saludar_completo("Miguel", "Hola"))
```

```python
# Funciones lambda
cuadrado = lambda x: x ** 2
print(f"Cuadrado de 5: {cuadrado(5)}")

# Usar con map y filter
numeros = [1, 2, 3, 4, 5]
cuadrados = list(map(lambda x: x ** 2, numeros))
pares = list(filter(lambda x: x % 2 == 0, numeros))

print(f"N칰meros: {numeros}")
print(f"Cuadrados: {cuadrados}")
print(f"Pares: {pares}")
```

## 6. List Comprehensions


```python
# List comprehension b치sico
cuadrados = [i ** 2 for i in range(10)]
print(f"Cuadrados: {cuadrados}")

# Con condici칩n
pares = [i for i in range(20) if i % 2 == 0]
print(f"Pares: {pares}")

# Dictionary comprehension
cuadrados_dict = {i: i ** 2 for i in range(5)}
print(f"Cuadrados dict: {cuadrados_dict}")
```

## 8. Programaci칩n Orientada a Objetos


```python
# Clase b치sica
class Persona:
    especie = "Homo sapiens"  # Atributo de clase

    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def presentarse(self):
        return f"Soy {self.nombre} y tengo {self.edad} a침os"

    @staticmethod
    def es_adulto(edad):
        return edad >= 18

# Crear instancia
persona1 = Persona("Miguel", 30)
print(persona1.presentarse())
print(f"쮼s adulto? {Persona.es_adulto(30)}")
```

```python
# Herencia
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def hacer_sonido(self):
        return "Hace un sonido"

class Perro(Animal):
    def hacer_sonido(self):
        return "Guau!"

    def ladrar(self):
        return f"{self.nombre} est치 ladrando"

perro = Perro("Fido")
print(perro.hacer_sonido())
print(perro.ladrar())
```

## 9. Ejercicios Pr치cticos


```python
# Ejercicio 1: Calculadora Simple
def calculadora(a, b, operacion):
    """Calcula operaci칩n entre dos n칰meros"""
    if operacion == '+':
        return a + b
    elif operacion == '-':
        return a - b
    elif operacion == '*':
        return a * b
    elif operacion == '/':
        if b == 0:
            return "Error: Divisi칩n por cero"
        return a / b
    else:
        return "Operaci칩n no v치lida"

# Probar
print(f"10 + 5 = {calculadora(10, 5, '+')}")
print(f"10 - 5 = {calculadora(10, 5, '-')}")
print(f"10 * 5 = {calculadora(10, 5, '*')}")
print(f"10 / 5 = {calculadora(10, 5, '/')}")
```

```python
# Ejercicio 2: Contador de Palabras
def contar_palabras(texto):
    """Cuenta la frecuencia de cada palabra en un texto"""
    palabras = texto.lower().split()
    frecuencia = {}
    for palabra in palabras:
        # Limpiar signos de puntuaci칩n
        palabra = palabra.strip('.,!?;:')
        frecuencia[palabra] = frecuencia.get(palabra, 0) + 1
    return frecuencia

texto = "Python es genial. Python es f치cil. Python es poderoso."
resultado = contar_palabras(texto)
print("Frecuencia de palabras:")
for palabra, count in resultado.items():
    print(f"  {palabra}: {count}")
```

```python
# Ejercicio 3: Filtrar Lista
def filtrar_numeros(numeros):
    """Filtra n칰meros pares mayores a 10"""
    return [n for n in numeros if n > 10 and n % 2 == 0]

numeros = [5, 12, 8, 15, 20, 3, 18, 7]
filtrados = filtrar_numeros(numeros)
print(f"N칰meros originales: {numeros}")
print(f"Pares mayores a 10: {filtrados}")
```

```python
# Ejercicio 4: Clase Calculadora con Historial
class Calculadora:
    def __init__(self):
        self.historial = []

    def sumar(self, a, b):
        resultado = a + b
        self.historial.append(f"{a} + {b} = {resultado}")
        return resultado

    def restar(self, a, b):
        resultado = a - b
        self.historial.append(f"{a} - {b} = {resultado}")
        return resultado

    def multiplicar(self, a, b):
        resultado = a * b
        self.historial.append(f"{a} * {b} = {resultado}")
        return resultado

    def dividir(self, a, b):
        if b == 0:
            raise ValueError("No se puede dividir por cero")
        resultado = a / b
        self.historial.append(f"{a} / {b} = {resultado}")
        return resultado

    def mostrar_historial(self):
        return self.historial

# Probar
calc = Calculadora()
calc.sumar(10, 5)
calc.restar(10, 3)
calc.multiplicar(4, 7)
calc.dividir(20, 4)

print("Historial de operaciones:")
for operacion in calc.mostrar_historial():
    print(f"  {operacion}")
```
