## 5. Funciones


```python
# Argumentos posicionales y por nombre
def calcular_area(base, altura):
    return base * altura

# Argumentos posicionales
area1 = calcular_area(5, 10)
print(f"calcular_area(5, 10) = {area1}")

# Argumentos por nombre
area2 = calcular_area(altura=10, base=5)
print(f"calcular_area(altura=10, base=5) = {area2}")

# Mezcla (posicionales primero, luego por nombre)
area3 = calcular_area(5, altura=10)
print(f"calcular_area(5, altura=10) = {area3}")
```

```python
# Función simple
def saludar():
    print("¡Hola!")

saludar()

# Función con parámetros
def saludar_nombre(nombre):
    print(f"¡Hola, {nombre}!")

saludar_nombre("Miguel")

# Función con valor por defecto
def saludar_completo(nombre, mensaje="Bienvenido"):
    print(f"{mensaje}, {nombre}!")

saludar_completo("Miguel")
saludar_completo("Miguel", "Hola")

# Función con retorno
def sumar(a, b):
    return a + b

resultado = sumar(5, 3)
print(f"\nsumar(5, 3) = {resultado}")

# Función con múltiples retornos
def dividir(a, b):
    cociente = a // b
    resto = a % b
    return cociente, resto

coc, res = dividir(10, 3)
print(f"dividir(10, 3) = cociente: {coc}, resto: {res}")
```

## 8. Programación Orientada a Objetos


```python
# Clase básica
class Persona:
    # Atributo de clase
    especie = "Homo sapiens"

    # Constructor
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    # Método de instancia
    def presentarse(self):
        return f"Soy {self.nombre} y tengo {self.edad} años"

    # Método estático
    @staticmethod
    def es_adulto(edad):
        return edad >= 18

    # Método de clase
    @classmethod
    def desde_ano_nacimiento(cls, nombre, ano_nacimiento):
        edad = 2024 - ano_nacimiento
        return cls(nombre, edad)

# Crear instancia
persona1 = Persona("Miguel", 30)
print(persona1.presentarse())

# Usar método estático
print(f"¿Es adulto? {Persona.es_adulto(25)}")

# Usar método de clase
persona2 = Persona.desde_ano_nacimiento("Ana", 1995)
print(persona2.presentarse())
```

```python
# Herencia
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def hacer_sonido(self):
        return "Hace un sonido"

class Perro(Animal):
    def hacer_sonido(self):
        return "Guau!"

    def ladrar(self):
        return f"{self.nombre} está ladrando"

perro = Perro("Fido")
print(perro.hacer_sonido())
print(perro.ladrar())

# Encapsulación
class CuentaBancaria:
    def __init__(self, saldo_inicial):
        self._saldo = saldo_inicial  # Protegido (convención)
        self.__pin = "1234"          # Privado (name mangling)

    def depositar(self, cantidad):
        if cantidad > 0:
            self._saldo += cantidad

    def obtener_saldo(self):
        return self._saldo

cuenta = CuentaBancaria(1000)
cuenta.depositar(500)
print(f"\nSaldo: {cuenta.obtener_saldo()}")
```

## 9. List Comprehensions


```python
# Forma tradicional vs List comprehension
# Forma tradicional
cuadrados_tradicional = []
for i in range(10):
    cuadrados_tradicional.append(i ** 2)

# List comprehension
cuadrados = [i ** 2 for i in range(10)]
print(f"Cuadrados (list comprehension): {cuadrados}")

# Con condición
pares = [i for i in range(20) if i % 2 == 0]
print(f"Pares: {pares}")

# Anidado
matriz = [[i * j for j in range(3)] for i in range(3)]
print(f"\nMatriz (anidado): {matriz}")

# Dictionary comprehension
cuadrados_dict = {i: i ** 2 for i in range(5)}
print(f"\nCuadrados dict: {cuadrados_dict}")

# Set comprehension
cuadrados_set = {i ** 2 for i in range(5)}
print(f"Cuadrados set: {cuadrados_set}")
```

```python
# Ejercicio 1: Calculadora Simple
def calculadora(a, b, operacion):
    """Calcula operación entre dos números"""
    if operacion == '+':
        return a + b
    elif operacion == '-':
        return a - b
    elif operacion == '*':
        return a * b
    elif operacion == '/':
        if b == 0:
            return "Error: División por cero"
        return a / b
    else:
        return "Operación no válida"

# Probar
print(f"10 + 5 = {calculadora(10, 5, '+')}")
print(f"10 - 5 = {calculadora(10, 5, '-')}")
print(f"10 * 5 = {calculadora(10, 5, '*')}")
print(f"10 / 5 = {calculadora(10, 5, '/')}")
```

```python
# Ejercicio 2: Clase Calculadora con Historial
class Calculadora:
    def __init__(self):
        self.historial = []

    def sumar(self, a, b):
        resultado = a + b
        self.historial.append(f"{a} + {b} = {resultado}")
        return resultado

    def restar(self, a, b):
        resultado = a - b
        self.historial.append(f"{a} - {b} = {resultado}")
        return resultado

    def multiplicar(self, a, b):
        resultado = a * b
        self.historial.append(f"{a} * {b} = {resultado}")
        return resultado

    def dividir(self, a, b):
        if b == 0:
            raise ValueError("No se puede dividir por cero")
        resultado = a / b
        self.historial.append(f"{a} / {b} = {resultado}")
        return resultado

    def mostrar_historial(self):
        return self.historial

# Probar
calc = Calculadora()
calc.sumar(10, 5)
calc.restar(10, 3)
calc.multiplicar(4, 7)
calc.dividir(20, 4)

print("Historial de operaciones:")
for operacion in calc.mostrar_historial():
    print(f"  {operacion}")
```

```python
# Ejercicio 4: Leer y Procesar CSV
import csv

# Leer CSV y calcular promedio de una columna
def calcular_promedio_columna(archivo_csv, indice_columna):
    """Calcula el promedio de una columna numérica en un CSV"""
    valores = []
    with open(archivo_csv, "r") as archivo:
        lector = csv.reader(archivo)
        encabezados = next(lector)  # Saltar encabezados
        for fila in lector:
            try:
                valor = float(fila[indice_columna])
                valores.append(valor)
            except (ValueError, IndexError):
                continue

    if valores:
        promedio = sum(valores) / len(valores)
        return promedio, valores
    return None, []

# Probar con el CSV creado anteriormente
promedio, valores = calcular_promedio_columna("datos_ejemplo.csv", 1)  # Columna de edad
if promedio:
    print(f"Promedio de edades: {promedio:.2f}")
    print(f"Valores: {valores}")
else:
    print("No se pudieron calcular promedios")
```
